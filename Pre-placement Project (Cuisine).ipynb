Data dictionary:

1. User ID: This is a unique identifier assigned to a specific user. 


2. Area Code: This is a numeric code used to identify a specific geographic area within a country. 


3. Location: This refers to the physical or geographic place where someone is located or resides.


4. Gender: Whether the customer is male or female


5. YOB: YOB stands for "Year of Birth." It indicates the year in which a person was born.


6. Marital Status: This refers to the current legal status of a person in relation to marriage.


7. Activity: This refers to the Profession the person has.


8. Budget: This refers to the amount of money set aside or available for spending.


9. Cuisines: This refers to the particular types of food or cooking styles associated with different cultures or regions. 


10. Alcohol: This refers to the consumption or preference for alcoholic beverages. 


11. Smoker: This indicates whether a person engages in the habit of smoking tobacco products or not.


12. Food Rating: This refers to the rating or evaluation given to a specific food item or meal.


13. Service Rating: This refers to the rating or evaluation given to the service provided by a restaurant, hotel, or any other establishment. 


14. Overall Rating: This refers to the overall evaluation or rating of a specific restaurant, hotel, or establishment. 


15. Often A S: Whether the customer visits the restaurant regularly or not

```python
### Segregating the values in categorical and continous columns
def seprate_data_types(df):
    categorical = []
    continuous = []
    for column in df.columns:                
        if df[column].nunique() < 40:      
                                                    
            categorical.append(column)
        else:
            continuous.append(column)
            
    return categorical, continuous 


categorical, continuous = seprate_data_types(df)         


from tabulate import tabulate
table = [categorical, continuous]
print(tabulate({"Categorical":categorical,
                "continuous": continuous}, headers = ["categorical", "continuous"]))
```

#### Bivariate Analysis

#### How does cuisine rating vary by location?

```python
location_ratings = df.groupby('Location')['Food Rating'].mean().reset_index()

# Sort the locations by average rating in descending order
location_ratings = location_ratings.sort_values(by='Food Rating', ascending=False)

# Create a bar chart to visualize the variation in cuisine ratings by location
plt.figure(figsize=(12, 6))
plt.bar(location_ratings['Location'], location_ratings['Food Rating'], color='skyblue')
plt.xlabel('Location')
plt.ylabel('Average Food Rating')
plt.title('Cuisine Ratings by Location')
plt.xticks(rotation=45)
plt.show()
```

##### Interpretation: The above chart provides the insights into how cuisine ratings vary based on the customers locations.Majority of the customers from Market City, NY gave 4.0 as average food rating.
