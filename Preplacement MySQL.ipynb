```python
import pandas as pd
from sqlalchemy import create_engine

# Credentials to database connection
hostname = "localhost"
dbname = "cuisine_rating"  # Name of the database into which you want to insert the data
uname = "root"
pwd = "new_password"

# Create a DataFrame
df = pd.read_csv("Cuisine_ratings.csv")

# Create SQLAlchemy engine to connect to MySQL Database
engine = create_engine(f"mysql+pymysql://{uname}:{pwd}@{hostname}/{dbname}")

# Convert DataFrame to a SQL table
df.to_sql('Customerdata', engine, if_exists='replace', index=False)

# Establish a database connection and execute a query
connection = engine.connect()
result = connection.execute("SELECT * FROM Customerdata").fetchall()

# Close the database connection
connection.close()
```
